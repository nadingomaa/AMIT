```python
from skimage.io import imread,imshow
import numpy as np
from skimage import data,feature
from skimage.morphology import dilation,erosion,closing,opening,disk, square, star,skeletonize
from skimage.util import invert
```

```python
im5=imread('./j.png')
```

```python
im5.shape
```

```python
im5=im5[:,:,0]
```

```python
imshow(im5)
```

**Dilation**

```python
imshow(dilation(image=im5,footprint=square(10)))
```

**Erosion**

```python
imshow(
    erosion(
        image = im5,
        footprint = disk(5)
    )
)
```

**Closing**

```python
im6=imread('./mri.png')
im6.shape
```

```python
imshow(im6)
```

**closing with dilation filter then erosion filter**

```python
im_dil=dilation(image=im6[:,:,0],footprint=disk(10))
imshow(im_dil)
```

```python
im_ero=erosion(image=im_dil,footprint=disk(10))
imshow(im_ero)
```

closing with closing filter

```python
close_im= closing(image = im6[:,:,0],footprint = disk(10))
imshow(close_im)
```

**opening with dilation filter then erosion filter**

```python
im_ero=erosion(image=im6[:,:,0],footprint=disk(10))
imshow(im_ero)
```

```python
im_dil=dilation(image=im_ero,footprint=disk(10))
imshow(im_dil)
```

opening with opening filter

```python
imshow(opening(image = im6[:,:,0],footprint = disk(10)))
```

**Skeletonization with Skeletonization filter**

```python
imshow(data.horse())
```

```python
# Invert the horse image
image = invert(data.horse())

# perform skeletonization
skeleton = skeletonize(image)
```

```python
imshow(skeleton)
```
